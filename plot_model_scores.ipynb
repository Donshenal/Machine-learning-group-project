## Make plot to compare models

```python
import seaborn as sns
import pandas as pd
```

```python
rmse_score = {
    'Linear' : 3.6,
    'Ridge' : 3.7,
    'Lasso' : 3.6,
    'KNN' : 2.5,
    'SVM' : 4.1,
    'XGBoost' : 1.9,
    'CatBoost' : 1.6
}
```

```python
df = pd.DataFrame(rmse_score.items(), columns=['Model', 'RMSE'])
```

```python
sns.barplot(data=df.sort_values(by='RMSE', ascending=False), x='Model', y='RMSE')
```
